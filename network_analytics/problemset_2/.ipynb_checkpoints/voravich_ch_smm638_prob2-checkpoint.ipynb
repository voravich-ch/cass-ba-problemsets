# Task
Use NumPy to implement the Girvan-Newman community detection algorithm.


## Girvan-Newman divisive procedure

- **step 1:** compute the centrality $x_{ij}$ of each link.
- **step 2:** remove the link with the largest centrality
- **step 3:** recalculate the centrality of each link for the altered network.
- **step 4:** iterate over steps 2 and 3 until all links are removed.

---
Notes: a good candidate of centrality is 'link betweenness', which captures the role of each link in information transfer. Hence $x_{ij}$ is proportional to the number of shortest paths between all node pairs that run along the link $(i,j)$.

## Girvan-Newman Community Detection Algorithm

```python
def gn_community_detection(G):
    
    # get a tuple containings nodes in each community as a dictionary
    community = tuple(nx.connected_components(G))
    
    # get the number of community
    num_community = len(community)
    
    # compute modularity
    modularity = nx.algorithms.community.quality.modularity(G, community)
    
    # store values as a list of tuples for m-n plot (modularity-num_community)
    store_mn = [(modularity, num_community)]
    
    """ Start using Girvan-Newman algorithm to remove edges"""
    # use while loop to iterate until the number of edges = 0
    while len(G.edges()) > 0:
        
        # get the dictionary that the keys: edge pairs; values: betweenness centrality
        betweenness = nx.edge_betweenness_centrality(G)
        
        # get the value of the maximum betweenness
        max_betweenness = max(betweenness.values())
        
        # get the index of the maximum betweenness
        idx_max_betweenness = list(betweenness.values()).index(max_betweenness)
        
        # get the key of that index: this is the edge with the largest centrality we want to remove
        edge_to_remove = list(betweenness.keys())[idx_max_betweenness]
        
        # remove that edge from the graph
        G.remove_edge(edge_to_remove[0], edge_to_remove[1])
        
        if len(G.edges()) > 0:

            # get a tuple containings nodes in each community as a dictionary
            community = tuple(nx.connected_components(G))

            # get the number of community
            new_num_community = len(community)

            # get the modularity for the new community structure
            new_modularity = nx.algorithms.community.quality.modularity(G, community)

            # append m,n to store_mn
            store_mn.append((new_modularity, new_num_community))

            if new_modularity > modularity:
                # save the graph that has better community structure as `optimal_G`
                # G.copy() function is needed to make a deep copy so that changes made to `G` do not affect `optimal_G`. 
                optimal_G = G.copy()
                # update modularity
                modularity = new_modularity
    
    return optimal_G, store_mn        
```

```python
""" 
The algorithm using max modularity is highly sensitive to noises 
so I decided to implement it on a simple example
"""
# l: groups, k: size 
G = nx.connected_caveman_graph(l = 6, k = 5)
```

```python
# plot decoration
options = {
    'node_color': 'purple',
    'alpha': 1,
    'width': 1,
    'with_labels': True,
    'font_color': 'white'}
pos = nx.kamada_kawai_layout(G)
```

```python
nx.draw(G, pos = pos, **options)
```

```python
optimal_G, store_mn = gn_community_detection(G)
```

```python
# plot decoration

# map the color of each node with its community
colors = ['orange', 'purple', 'tomato', 'cornflowerblue', 'yellowgreen', 'hotpink']
color_map = []
# loop over nodes in G
for node in G:
    # loop over number of communities
    for i in range(len(tuple(nx.connected_components(optimal_G)))):
        if node in tuple(nx.connected_components(optimal_G))[i]:
            color_map.append(colors[i])

options = {
    'node_color': color_map,
    'alpha': 1,
    'width': 1,
    'with_labels': True,
    'font_color': 'white'}
```

```python
nx.draw(optimal_G, pos = pos, **options)
```

```python
# set plot style
plt.style.use('fivethirtyeight')
# unzip store_mn
y, x = zip(*store_mn)
# add vertical line to the number of communities the algorithm picked
plt.plot(x, y, linewidth = 2)
plt.axvline(x = len(tuple(nx.connected_components(optimal_G))), color = 'red', linewidth = 1.5, ls = ':')
plt.ylabel('Modularity')
plt.xlabel('Number of Communities')
plt.title('Model Selection')
plt.show()
```

# Additional: More conservative Girvan-Newman Community Detection Algorithm

```python
def gn_community_detection(G):
    
    # get a tuple containings nodes in each community as a dictionary
    community = tuple(nx.connected_components(G))
    
    # get the number of community
    num_community = len(community)
    
    # compute modularity
    modularity = nx.algorithms.community.quality.modularity(G, community)
    
    # store values as a list of tuples for m-n plot (modularity-num_community)
    store_mn = [(modularity, num_community)]
    
    """ Start using Girvan-Newman algorithm to remove edges"""
    # use while loop to iterate until the number of edges = 0
    while len(G.edges()) > 0:
        # get the dictionary that the keys: edge pairs; values: betweenness centrality
        betweenness = nx.edge_betweenness_centrality(G)
        
        # get the value of the maximum betweenness
        max_betweenness = max(betweenness.values())
        
        # get the index of the maximum betweenness
        idx_max_betweenness = list(betweenness.values()).index(max_betweenness)
        
        # get the key of that index: this is the edge with the largest centrality we want to remove
        edge_to_remove = list(betweenness.keys())[idx_max_betweenness]
        
        # remove that edge from the graph
        G.remove_edge(edge_to_remove[0], edge_to_remove[1])
        
        if len(G.edges()) > 0:

            # get a tuple containings nodes in each community as a dictionary
            community = tuple(nx.connected_components(G))

            # get the number of community
            new_num_community = len(community)

            # get the modularity for the new community structure
            new_modularity = nx.algorithms.community.quality.modularity(G, community)

            # append m,n to store_mn
            store_mn.append((new_modularity, new_num_community))
            
            """ {CHANGES}: only save the model if the information gained (modularity) 
            increased by 10% from the previous one """
            if new_modularity - modularity > 0.10:
                # save the graph that has better community structure as `optimal_G`
                # G.copy() function is needed to make a deep copy so that changes made to `G` do not affect `optimal_G`. 
                optimal_G = G.copy()
                # update modularity
            modularity = new_modularity
    
    return optimal_G, store_mn        
```
